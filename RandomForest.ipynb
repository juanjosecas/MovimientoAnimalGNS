## Normalización y Cálculos de Distancias

### 1. Normalización de Coordenadas
La normalización de las coordenadas se realiza para que estén en un rango [0, 1], lo que facilita el análisis comparativo y la interpretación de los datos. Las coordenadas (x, y) de cada punto se normalizan utilizando las dimensiones de la imagen:

$$
\text{Coordenada Normalizada} = \frac{\text{Coordenada Original}}{\text{Dimensión}}
$$

Donde:
- $\text{Dimensión}$ es el ancho o alto del video, según corresponda.

### 2. Cálculo de Distancias
Para calcular la distancia euclidiana entre dos puntos $(x_1, y_1)$ y $(x_2, y_2)$, utilizamos la siguiente fórmula:

$$
d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
$$

Esta distancia se utiliza para calcular:
- La distancia desde la nariz del ratón hasta el hisopo.
- La distancia desde la cabeza hasta la base del ratón.

### 3. Cálculo de Ángulos
Los ángulos formados por los puntos del ratón se calculan utilizando el producto punto y la norma de los vectores:

$$
\theta = \arccos\left(\frac{\vec{v_1} \cdot \vec{v_2}}{||\vec{v_1}|| \cdot ||\vec{v_2}||}\right)
$$

Donde:
- $\vec{v_1}$ es el vector desde el lomo hasta la cabeza.
- $\vec{v_2}$ es el vector desde el lomo hasta la base.

Esto se convierte en grados usando:

$$
\theta_{\text{grados}} = \frac{\theta \cdot 180}{\pi}
$$

### 4. Cálculo de Velocidades y Aceleraciones
Las velocidades se calculan a partir de las distancias utilizando el tiempo como ventana de observación. Para calcular la velocidad de la nariz, utilizamos la siguiente fórmula:

$$
\text{Velocidad} = \frac{d(t) - d(t - \text{ventana})}{\text{ventana}}
$$

Donde $d(t)$ es la distancia en el tiempo $t$ y $\text{ventana}$ es el tamaño de la ventana en frames.

La aceleración se calcula como la diferencia entre las velocidades en diferentes frames:

$$
\text{Aceleración} = \frac{\text{Velocidad}(t) - \text{Velocidad}(t - \text{ventana})}{\text{ventana}}
$$

### 5. Normalización de Ángulos
Los ángulos se normalizan para estar en un rango [0, 1] dividiendo por 180 grados:

$$
\text{Ángulo Normalizado} = \frac{\text{Ángulo Original}}{180}
$$

### 6. Normalización de Distancias
Las distancias calculadas se normalizan utilizando la diagonal de la imagen:

$$
\text{Distancia Normalizada} = \frac{\text{Distancia Original}}{\text{Diagonal}}
$$

Donde $\text{Diagonal}$ se calcula como:

$$
\text{Diagonal} = \sqrt{\text{ancho}^2 + \text{alto}^2}
$$

## Evaluación del Balanceo de Observaciones

La función `evaluar_balanceo` se utiliza para analizar el balance de clases en el conjunto de datos de observaciones. Un conjunto de datos equilibrado es fundamental para evitar sesgos en el modelo de clasificación. A continuación se describen las métricas calculadas en esta función:

### 1. Conteo de Observaciones
Primero, contamos el número de ocurrencias de cada clase en la serie de observaciones:

$$
\text{counts} = \text{observations.value\_counts()}
$$

Esto genera una serie donde cada clase se asocia con su cantidad de ocurrencias.

### 2. Cálculo del Índice de Gini
El índice de Gini es una medida de desigualdad en la distribución de las clases. Se calcula como:

$$
\text{Gini Index} = 1 - \sum \left(\frac{count_i}{total}\right)^2
$$

Donde:
- $ count_i $ es el conteo de la clase $ i $.
- $ total $ es el número total de observaciones.

Un índice de Gini más bajo indica un mayor desequilibrio.

### 3. Cálculo del Índice de Shannon
El índice de Shannon, o entropía, mide la incertidumbre en la distribución de clases. Se calcula usando la función `entropy` de SciPy:

$$
\text{Shannon Index} = -\sum \left(\frac{count_i}{total} \cdot \log\left(\frac{count_i}{total}\right)\right)
$$

Un índice de Shannon más alto indica una mayor diversidad en la distribución de clases.

### 4. Relación Mínimo-Máximo
La relación mínimo-máximo se calcula como:

$$
\text{Min-Max Ratio} = \frac{\text{min(counts)}}{\text{max(counts)}}
$$

Esto proporciona una medida rápida del balanceo entre la clase menos representada y la más representada. Un valor más alto indica un mayor balance.

### 5. Prueba Chi-Cuadrado
La prueba de Chi-Cuadrado se utiliza para evaluar la independencia de las clases. Se calcula usando la función `chi2_contingency` de SciPy, que devuelve dos valores: el estadístico Chi-Cuadrado y el valor p asociado. El valor p indica si hay una diferencia estadísticamente significativa entre las clases:

$$
\text{Chi2 p-value}
$$

Un valor p mayor a 0.05 sugiere que no hay evidencia suficiente para rechazar la hipótesis nula de que las clases están distribuidas uniformemente.

### 6. Evaluación del Balanceo
Finalmente, se determina si el conjunto de datos está balanceado utilizando las siguientes condiciones:

$$
\text{is\_balanced} = \text{min\_max\_ratio} > 0.5 \text{ y } \text{gini\_index} > 0.5 \text{ y } \text{p\_value} > 0.05
$$

Si todas estas condiciones se cumplen, se considera que el conjunto de datos está equilibrado.

````python
# Importamos las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import FunctionTransformer
from scipy.stats import entropy, chi2_contingency

# Definir las dimensiones de la imagen
width = 520  # Ancho del video en píxeles
height = 245  # Alto del video en píxeles
hisopo_x = 270
hisopo_y = 50
frames_habituacion = 9000

diagonal = np.sqrt(width**2 + height**2)  # Diagonal de la imagen para la normalización

# Función para normalizar las coordenadas en función del tamaño del video
def normalizar_coordenadas(df):
    cols_x = [col for col in df.columns if '.x' in col]
    cols_y = [col for col in df.columns if '.y' in col]
    df[cols_x] = df[cols_x] / width
    df[cols_y] = df[cols_y] / height
    return df

# Función para calcular la distancia euclidiana entre dos puntos (x1, y1) y (x2, y2)
def calcular_distancia(x1, y1, x2, y2):
    return np.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

# Calcular el ángulo formado por los puntos: Cabeza - Lomo - Base
def calcular_angulo(df):
    vector1 = np.column_stack((df['Cabeza.x'] - df['Lomo.x'], df['Cabeza.y'] - df['Lomo.y']))
    vector2 = np.column_stack((df['Base.x'] - df['Lomo.x'], df['Base.y'] - df['Lomo.y']))
    dot_product = np.einsum('ij,ij->i', vector1, vector2)
    norm1 = np.linalg.norm(vector1, axis=1)
    norm2 = np.linalg.norm(vector2, axis=1)
    angles = np.arccos(dot_product / (norm1 * norm2))
    return np.degrees(angles)  # Convertir a grados

# Función para calcular el ángulo de orientación hacia el hisopo
def calcular_angulo_hisopo(df):
    vector_cuerpo = np.column_stack((df['Base.x'] - df['Cabeza.x'], df['Base.y'] - df['Cabeza.y']))
    vector_hisopo = np.column_stack((hisopo_x - df['Lomo.x'], hisopo_y - df['Lomo.y']))
    dot_product = np.einsum('ij,ij->i', vector_cuerpo, vector_hisopo)
    norm_cuerpo = np.linalg.norm(vector_cuerpo, axis=1)
    norm_hisopo = np.linalg.norm(vector_hisopo, axis=1)
    ang_hisopo = np.arccos(dot_product / (norm_cuerpo * norm_hisopo))
    return np.degrees(ang_hisopo)  # Convertir a grados

# Función para normalizar los ángulos en un rango [0, 1]
def normalizar_angulos(df):
    cols_angulos = ['ang_cuerpo', 'ang_hisopo']
    for col in cols_angulos:
        df[col] = df[col] / 180  # Normaliza dividiendo por 180 grados
    return df

# Función para calcular velocidades y aceleraciones de los puntos y ángulos
def calcular_vel_acel(df, window_size=60):
    df['vel_nariz'] = calcular_distancia(
        df['Nariz.x'].shift(window_size), df['Nariz.y'].shift(window_size), df['Nariz.x'], df['Nariz.y']
    ) / window_size
    df['acel_nariz'] = df['vel_nariz'].diff(window_size) / window_size
    df['vel_ang_cuerpo'] = df['ang_cuerpo'].diff(window_size) / window_size
    df['vel_ang_hisopo'] = df['ang_hisopo'].diff(window_size) / window_size
    df[['vel_nariz', 'acel_nariz', 'vel_ang_cuerpo', 'vel_ang_hisopo']] = df[
        ['vel_nariz', 'acel_nariz', 'vel_ang_cuerpo', 'vel_ang_hisopo']
    ].fillna(0)
    return df

# Función para normalizar distancias en función de la diagonal de la imagen
def normalizar_distancias(df):
    cols_distancias = ['nariz_hisopo', 'cabeza_hisopo', 'long_cuerpo']
    for col in cols_distancias:
        df[col] = df[col] / diagonal
    return df

# Función para extraer características del comportamiento del ratón
def extraer_caracteristicas(df, window_size=1):
    df['nariz_hisopo'] = calcular_distancia(df['Nariz.x'], df['Nariz.y'], hisopo_x, hisopo_y)
    df['cabeza_hisopo'] = calcular_distancia(df['Cabeza.x'], df['Cabeza.y'], hisopo_x, hisopo_y)
    df['long_cuerpo'] = calcular_distancia(df['Cabeza.x'], df['Cabeza.y'], df['Base.x'], df['Base.y'])
    df['ang_cuerpo'] = calcular_angulo(df)
    df['ang_hisopo'] = calcular_angulo_hisopo(df)
    df = normalizar_distancias(df)  # Normalizar distancias calculadas
    df = normalizar_angulos(df)  # Normalizar ángulos calculados
    df = calcular_vel_acel(df, window_size)
    df.dropna(inplace=True)
    return df

# Función para filtrar los primeros N frames de habituación y exploración
def filtrar_frames_habituacion(df, num_frames_habituacion=0):
    df_filtrado = df[df['frame_idx'] > num_frames_habituacion]
    return df_filtrado

# Aplicar las funciones para normalizar, extraer características y filtrar los primeros N frames
df = normalizar_coordenadas(df)
df = extraer_caracteristicas(df, window_size=5)  # Cambia el tamaño de la ventana según se requiera
df = filtrar_frames_habituacion(df, num_frames_habituacion=frames_habituacion)  # Cambia este valor según los frames de habituación que quieras filtrar

# Mostrar una muestra del DataFrame resultante para verificar que todo esté correcto
print(df.head())

# Estadísticas básicas de las observaciones para ver el balance de las clases
observation_counts = df['Observation'].value_counts()
print(observation_counts)

# Visualización del balance de las clases
observation_counts.plot(kind='bar', title='Distribución de Observaciones', ylabel='Cantidad', xlabel='Comportamiento')
plt.show()

# Evaluar el balanceo de las observaciones para ver si necesitamos aplicar alguna técnica de balanceo
def evaluar_balanceo(observations):
    counts = observations.value_counts()
    total = counts.sum()
    gini_index = 1 - sum((counts / total) ** 2)
    shannon_index = entropy(counts)
    min_max_ratio = counts.min() / counts.max()
    chi2, p_value = chi2_contingency([counts.values])[0:2]
    is_balanced = min_max_ratio > 0.5 and gini_index > 0.5 and p_value > 0.05
    return {
        "Gini Index": gini_index,
        "Shannon Index": shannon_index,
        "Min-Max Ratio": min_max_ratio,
        "Chi2 p-value": p_value,
        "Balanced": is_balanced
    }

resultado_balanceo = evaluar_balanceo(df['Observation'])
print(resultado_balanceo)
````

## Evaluación de Métodos de Balanceo de Clases

En esta sección, evaluamos diferentes métodos de balanceo de clases para mejorar el rendimiento de un modelo de clasificación en un conjunto de datos desequilibrado. Utilizamos el clasificador Random Forest para este propósito.

### 1. División de los Datos
Primero, separamos los datos en características ($X$) y la variable objetivo ($y$). Luego, dividimos el conjunto de datos en un conjunto de entrenamiento y otro de prueba:

- $X = df.drop(columns=['Observation', 'frame_idx'])$
- $y = df['Observation']$

La división se realiza usando `train_test_split`, asegurando que la proporción de clases se mantenga (usando el argumento `stratify`):

$$
\begin{align*}
X_{\text{train}}, X_{\text{test}}, y_{\text{train}}, y_{\text{test}} & = \text{train\_test\_split}(X, y, \text{test\_size}=0.1, \text{random\_state}=42, \text{stratify}=y)
\end{align*}
$$

### 2. Métodos de Balanceo
Definimos varios métodos de balanceo a evaluar. En este caso, utilizamos el submuestreo aleatorio y el método SMOTE seguido de ENN (Edited Nearest Neighbors):

- **Random UnderSampler**: Reduce el número de ejemplos de la clase mayoritaria.
- **SMOTEENN**: Genera ejemplos sintéticos y luego elimina ejemplos de las clases cercanas.

### 3. Evaluación de Métodos de Balanceo
Para cada método de balanceo definido, realizamos los siguientes pasos:

- Aplicamos el método de balanceo al conjunto de entrenamiento.
- Definimos y entrenamos un modelo de Random Forest con el argumento `class_weight='balanced'` para manejar la desproporción en las clases.
- Realizamos validación cruzada para evaluar la efectividad del método utilizando el F1-Score macro, que es útil para evaluar el rendimiento en conjuntos de datos desbalanceados.

$$
\text{F1-Score} = \frac{2 \cdot \text{Precisión} \cdot \text{Recall}}{\text{Precisión} + \text{Recall}}
$$

Los resultados se almacenan y se imprime el F1-Score promedio y su desviación estándar para cada método de balanceo.

### 4. Selección del Mejor Método
Después de evaluar todos los métodos, seleccionamos el que proporciona el mejor F1-Score promedio:

$$
\text{mejor\_metodo} = \max(\text{resultados}, \text{key}=\text{resultados.get})
$$

### 5. Entrenamiento del Modelo Final
Aplicamos el mejor método de balanceo al conjunto completo de entrenamiento y ajustamos el modelo final de Random Forest con los datos balanceados. Luego, evaluamos el modelo final en el conjunto de prueba y mostramos un reporte de clasificación que incluye métricas como precisión, recall y F1-score.

$$
\text{y\_pred} = \text{modelo\_final.predict}(X_{\text{test}})
$$


````python
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTEENN, SMOTETomek
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# División inicial de los datos
X = df.drop(columns=['Observation', 'frame_idx'])
y = df['Observation']

# División en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)

# Definición de los métodos de balanceo a evaluar
metodos_balanceo = {
    'Undersampling': RandomUnderSampler(random_state=42),
    'SMOTEENN': SMOTEENN(random_state=42)
}

# Evaluar cada método de balanceo
resultados = {}
for nombre, metodo in metodos_balanceo.items():
    print(f"\nEvaluando método de balanceo: {nombre}")
    # Aplicar el método de balanceo al conjunto de entrenamiento
    X_res, y_res = metodo.fit_resample(X_train, y_train)
    
    # Definir el modelo a utilizar (Random Forest)
    modelo = RandomForestClassifier(random_state=42, class_weight='balanced')
    
    # Validación cruzada para evaluar la efectividad del método de balanceo
    scores = cross_val_score(modelo, X_res, y_res, cv=StratifiedKFold(n_splits=5), scoring='f1_macro')
    
    # Guardar resultados
    resultados[nombre] = scores.mean()
    print(f"F1-Score promedio: {scores.mean():.4f} (± {scores.std():.4f})")

# Selección del mejor método basado en el F1-score promedio
mejor_metodo = max(resultados, key=resultados.get)
print(f"\nMejor método de balanceo encontrado: {mejor_metodo} con F1-Score promedio de {resultados[mejor_metodo]:.4f}")

# Aplicar el mejor método de balanceo al conjunto completo de entrenamiento y ajustar el modelo final
mejor_balanceo = metodos_balanceo[mejor_metodo]
X_res, y_res = mejor_balanceo.fit_resample(X_train, y_train)
modelo_final = RandomForestClassifier(random_state=42, class_weight='balanced')
modelo_final.fit(X_res, y_res)

# Evaluar el modelo final en el conjunto de prueba
y_pred = modelo_final.predict(X_test)
print("\nReporte de Clasificación en Conjunto de Prueba:")
print(classification_report(y_test, y_pred))
````

## Balanceo de Datos y Evaluación

En esta sección, implementamos una función para balancear un DataFrame utilizando un método de balanceo especificado. El objetivo es abordar el problema de clases desbalanceadas en nuestro conjunto de datos, lo cual es crucial para mejorar el rendimiento del modelo de clasificación.

### 1. Función de Balanceo
La función `balancear` se encarga de equilibrar las clases en el DataFrame:

- **Entradas**:
  - `df`: DataFrame original que contiene datos desbalanceados.
  - `metodo_balanceo`: Método de balanceo a utilizar (por ejemplo, SMOTE, SMOTEENN).

- **Proceso**:
  - Separamos las características ($X$) de la variable objetivo ($y$):
    $$
    X = df.drop(columns=['Observation', 'frame\_idx'], \text{errors='ignore'})
    $$
    $$
    y = df['Observation']
    $$
  - Aplicamos el método de balanceo seleccionado:
    $$
    (X_{\text{balanced}}, y_{\text{balanced}}) = \text{metodo\_balanceo.fit\_resample}(X, y)
    $$
  - Creamos un nuevo DataFrame balanceado que combina $X_{\text{balanced}}$ y $y_{\text{balanced}}$:
    $$
    df_{\text{balanced}} = \text{pd.concat}([ \text{pd.DataFrame}(X_{\text{balanced}}, \text{columns}=X.columns), \text{pd.DataFrame}(y_{\text{balanced}}, \text{columns}=['Observation'])], \text{axis}=1)
    $$

### 2. Ejemplo de Uso
Se inicia el proceso de balanceo utilizando el método SMOTEENN. Se mide el tiempo que toma realizar el balanceo y se imprime el resultado.

### 3. Evaluación del Balanceo
Después de obtener el DataFrame balanceado, evaluamos la efectividad del balanceo utilizando la función `evaluar_balanceo`. Esta función calcula métricas que nos permiten verificar si el conjunto de datos ahora está equilibrado:

- **Índice de Gini**: Medida de desigualdad en la distribución de clases.
- **Índice de Shannon**: Mide la incertidumbre de la distribución de clases.
- **Relación Mínimo-Máximo**: Proporción entre la clase menos representada y la más representada.
- **Prueba Chi-Cuadrado**: Evaluación de la independencia de las clases.

El resultado se imprime para verificar si el balanceo fue exitoso y si se necesita aplicar alguna técnica adicional.

### Resultados
Finalmente, se muestran los resultados de la evaluación del balanceo, que incluyen el Índice de Gini, Índice de Shannon, Relación Mínimo-Máximo, p-valor de Chi-Cuadrado, y una indicación de si el conjunto de datos está balanceado.


````python
import pandas as pd
from imblearn.combine import SMOTEENN  # Asegúrate de importar el método de balanceo que funcionó mejor
import time
from scipy.stats import entropy, chi2_contingency

# Función para balancear el DataFrame usando el método de balanceo especificado
def balancear(df, metodo_balanceo):
    """
    Balancea el DataFrame usando el método especificado.
    
    :param df: DataFrame original con datos desbalanceados.
    :param metodo_balanceo: Método de balanceo a usar (ej. SMOTE, SMOTEENN, etc.).
    :return: DataFrame balanceado.
    """
    # Separar características y etiquetas
    X = df.drop(columns=['Observation', 'frame_idx'], errors='ignore')  # Remueve 'Observation' y 'frame_idx' si existen
    y = df['Observation']
    
    # Aplicar el método de balanceo seleccionado
    X_balanced, y_balanced = metodo_balanceo.fit_resample(X, y)
    
    # Crear el nuevo DataFrame balanceado
    df_balanced = pd.concat([pd.DataFrame(X_balanced, columns=X.columns), pd.DataFrame(y_balanced, columns=['Observation'])], axis=1)
    
    return df_balanced

# Ejemplo de uso de la función con SMOTEENN como método de balanceo
start_time = time.time()

print('Comienzo de balanceo')
metodo_balanceo = SMOTEENN(random_state=42)  # Puedes cambiar por otro método si has encontrado uno más efectivo
df_balanced = balancear(df, metodo_balanceo)

print(f'Fin balanceo en {int(time.time() - start_time)} segs')

# Mostrar las primeras filas del DataFrame balanceado
print(df_balanced.head())

# Evaluar el balanceo de las observaciones para ver si necesitamos aplicar alguna técnica de balanceo
def evaluar_balanceo(observations):
    counts = observations.value_counts()
    total = counts.sum()
    gini_index = 1 - sum((counts / total) ** 2)
    shannon_index = entropy(counts)
    min_max_ratio = counts.min() / counts.max()
    chi2, p_value = chi2_contingency([counts.values])[0:2]
    is_balanced = min_max_ratio > 0.5 and gini_index > 0.5 and p_value > 0.05
    return {
        "Gini Index": gini_index,
        "Shannon Index": shannon_index,
        "Min-Max Ratio": min_max_ratio,
        "Chi2 p-value": p_value,
        "Balanced": is_balanced
    }

# Evaluar y mostrar el resultado del balanceo
resultado_balanceo = evaluar_balanceo(df_balanced['Observation'])
print("\nResultados de la evaluación del balanceo:")
print(resultado_balanceo)
````

### Random Forest: Fundamentos y Aplicación

**¿Qué es Random Forest?**
Random Forest es un algoritmo de aprendizaje automático basado en la creación de múltiples árboles de decisión durante el entrenamiento y la predicción. Cada árbol se entrena sobre un subconjunto aleatorio de los datos y de características, lo que permite mejorar la generalización del modelo y reducir el riesgo de sobreajuste. La predicción final se obtiene mediante un voto de mayoría (para clasificación) o promediando (para regresión) los resultados de todos los árboles.

**¿Por qué usar Random Forest en este caso?**
- **Manejo de Datos Desbalanceados**: Random Forest puede manejar conjuntos de datos desbalanceados mediante la ponderación de clases, lo que permite dar mayor importancia a las clases minoritarias (como el "Evento" en este caso).
- **Robustez y Precisión**: Debido a su naturaleza de ensamblado, Random Forest tiende a ser más robusto y preciso que un único árbol de decisión, lo que lo hace adecuado para problemas complejos.
- **Interpretabilidad**: Aunque menos interpretables que un único árbol, los modelos de Random Forest pueden proporcionar información sobre la importancia de las características.

**Otros Métodos Sugeridos**:
- **Gradient Boosting**: Un enfoque similar que combina árboles de decisión, optimizando sucesivos árboles para corregir errores de modelos anteriores.
- **SVM (Support Vector Machines)**: Funciona bien en espacios de alta dimensión y puede ser eficaz para clasificación binaria.
- **k-Nearest Neighbors (k-NN)**: Un método simple que clasifica un punto según la mayoría de sus vecinos cercanos, útil en conjuntos de datos más pequeños.

### 1. Generación de Informe del Modelo

La función `guardar_graficos` genera gráficos para evaluar el rendimiento del modelo:

Al final del bloque, se ejecutan las funciones para generar el informe del modelo y guardar tanto el modelo entrenado como los gráficos generados.
````python
import json
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, roc_curve, roc_auc_score

# Función para generar un informe detallado del modelo entrenado y guardar gráficos
def generar_informe_modelo(modelo, X_test, y_test, feature_names, informe_path='informe_modelo.txt'):
    """
    Genera un informe detallado del modelo entrenado y guarda sus métricas, características e imágenes gráficas.

    :param modelo: El modelo de Random Forest entrenado.
    :param X_test: Conjunto de características de prueba escalado.
    :param y_test: Etiquetas verdaderas del conjunto de prueba.
    :param feature_names: Lista de nombres de las características utilizadas en el modelo.
    :param informe_path: Ruta donde se guardará el informe del modelo.
    """
    # Realizar predicciones
    y_pred = modelo.predict(X_test)
    y_prob = modelo.predict_proba(X_test)[:, 1]  # Probabilidad de la clase "Evento"

    # Calcular métricas del modelo
    accuracy = accuracy_score(y_test, y_pred)
    classification_rep = classification_report(y_test, y_pred, digits=5)
    confusion_mat = confusion_matrix(y_test, y_pred)
    auc_score = roc_auc_score(y_test, y_prob)

    # Estructura del informe
    informe = {
        'Modelo': str(type(modelo)),
        'Hiperparámetros': modelo.get_params(),
        'Métricas': {
            'Accuracy': round(accuracy, 5),
            'Reporte de Clasificación': classification_rep,
            'Matriz de Confusión': confusion_mat.tolist(),
            'AUC Score': round(auc_score, 5)
        },
        'Importancia de Características': {name: round(imp, 5) for name, imp in zip(feature_names, modelo.feature_importances_)},
        'Descripción': 'Modelo Random Forest optimizado y entrenado con datos balanceados, evaluado con métricas de precisión, '
                       'recall, F1-score, y AUC. Incluye una descripción de la importancia de cada característica.'
    }

    # Guardar el informe en un archivo JSON
    with open(informe_path, 'w') as file:
        json.dump(informe, file, indent=4)

    # También guardar el informe como texto para fácil lectura
    with open(informe_path.replace('.json', '.txt'), 'w') as file_txt:
        file_txt.write("=== INFORME DEL MODELO ENTRENADO ===\n")
        file_txt.write(f"Modelo: {type(modelo)}\n\n")
        file_txt.write("=== Hiperparámetros del Modelo ===\n")
        for param, value in modelo.get_params().items():
            file_txt.write(f"{param}: {value}\n")

        file_txt.write("\n=== Métricas del Modelo ===\n")
        file_txt.write(f"Accuracy: {accuracy:.5f}\n")
        file_txt.write(f"AUC Score: {auc_score:.5f}\n")
        file_txt.write(f"Reporte de Clasificación:\n{classification_rep}\n")
        file_txt.write(f"Matriz de Confusión:\n{confusion_mat}\n")

        file_txt.write("\n=== Importancia de las Características ===\n")
        for name, importance in zip(feature_names, modelo.feature_importances_):
            file_txt.write(f"{name}: {importance:.5f}\n")

    print(f"Informe guardado en: {informe_path}")

    # Guardar los gráficos generados
    guardar_graficos(y_test, y_pred, y_prob, modelo.classes_)

# Función para guardar gráficos de evaluación del modelo
def guardar_graficos(y_test, y_pred, y_prob, classes, plot_path_prefix='modelo_RandomForest'):
    """
    Guarda los gráficos de la matriz de confusión y la curva ROC para el modelo.

    :param y_test: Etiquetas verdaderas del conjunto de prueba.
    :param y_pred: Predicciones realizadas por el modelo.
    :param y_prob: Probabilidades de la clase "Evento".
    :param classes: Lista de clases del modelo.
    :param plot_path_prefix: Prefijo de ruta para guardar los gráficos.
    """
    # Matriz de Confusión
    plt.figure(figsize=(10, 7))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues', 
                xticklabels=classes, yticklabels=classes)
    plt.xlabel('Predicciones')
    plt.ylabel('Observaciones Reales')
    plt.title('Matriz de Confusión - Modelo Optimizado y Balanceado')
    plt.savefig(f'{plot_path_prefix}_matriz_confusion.png')
    plt.close()

    # Curva ROC
    fpr, tpr, thresholds = roc_curve(y_test, y_prob, pos_label="Evento")
    plt.figure(figsize=(10, 6))
    plt.plot(fpr, tpr, label=f'AUC = {roc_auc_score(y_test, y_prob):.5f}', color='darkorange')
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Curva ROC - Modelo Optimizado')
    plt.legend()
    plt.grid()
    plt.savefig(f'{plot_path_prefix}_curva_ROC.png')
    plt.close()

    print(f"Gráficos guardados con el prefijo: {plot_path_prefix}")

# Guardar el modelo entrenado para futuras predicciones o reentrenamiento
def guardar_modelo(modelo, model_path='modelo_rf_optimizado.joblib'):
    """
    Guarda el modelo entrenado en un archivo para reutilización o reentrenamiento.

    :param modelo: El modelo de Random Forest entrenado.
    :param model_path: Ruta donde se guardará el modelo.
    """
    joblib.dump(modelo, model_path)
    print(f"Modelo guardado en: {model_path}")

# Ejecución de las funciones para generar el informe, guardar los gráficos y el modelo
model_file_save = 'modelo_RandomForest_A7-B4-C3.joblib'
informe_path='informe_modelo_RandomForest_A7-B4-C3.json'

generar_informe_modelo(best_model_rf, X_test_scaled, y_test, df_reducido_final.columns, informe_path=informe_path)
guardar_modelo((best_model_rf, df_reducido_final.columns), model_path=model_file_save)
````

````python
import pandas as pd
import numpy as np
import joblib
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Función para cargar y predecir utilizando el archivo de salida de SLEAP
def cargar_y_predecir_sleap(model_path, sleap_csv_path, window_size=60, num_frames_habituacion=9000):
    """
    Carga un modelo entrenado y aplica predicciones a un archivo de salida de SLEAP.
    
    :param model_path: Ruta al archivo del modelo entrenado (.joblib).
    :param sleap_csv_path: Ruta al archivo CSV de salida de SLEAP a procesar.
    :param window_size: Tamaño de la ventana de tiempo para calcular velocidades y aceleraciones.
    :param num_frames_habituacion: Número de frames iniciales que corresponden a la fase de habituación.
    :return: DataFrame con las predicciones y probabilidades.
    """
    # Cargar el modelo entrenado y las características
    modelo, feature_names = joblib.load(model_path)
    print(f"Modelo cargado desde: {model_path}")

    # Cargar el archivo de salida de SLEAP
    df_sleap = pd.read_csv(sleap_csv_path)
    print(f"Archivo de SLEAP cargado desde: {sleap_csv_path}")

    # Rellenar NaN en coordenadas con un valor específico para evitar errores
    df_sleap.fillna(0, inplace=True)

    # Calcular las coordenadas de la Cabeza como promedio de Nariz, Frente, O_D y O_I
    df_sleap['Cabeza.x'] = df_sleap[['Nariz.x', 'Frente.x', 'O_D.x', 'O_I.x']].mean(axis=1)
    df_sleap['Cabeza.y'] = df_sleap[['Nariz.y', 'Frente.y', 'O_D.y', 'O_I.y']].mean(axis=1)

    # Aplicar el preprocesamiento: normalización de coordenadas y extracción de características
    df_sleap = normalizar_coordenadas(df_sleap)
    df_sleap = extraer_caracteristicas(df_sleap, window_size)
    df_sleap = filtrar_frames_habituacion(df_sleap, num_frames_habituacion)

    # Verificar que las columnas necesarias están presentes, ignorando 'Observation' si no está
    missing_features = [feat for feat in feature_names if feat not in df_sleap.columns]
    if missing_features and 'Observation' not in missing_features:
        raise ValueError(f"Las siguientes características necesarias no están presentes en el DataFrame: {missing_features}")

    # Seleccionar solo las características necesarias y en el orden correcto, ignorando 'Observation'
    X_sleap = df_sleap[[col for col in feature_names if col in df_sleap.columns]]

    # Normalizar las características usando StandardScaler
    scaler = StandardScaler()
    X_sleap_scaled = scaler.fit_transform(X_sleap)

    # Realizar predicciones con el modelo cargado
    y_pred = modelo.predict(X_sleap_scaled)
    y_prob = modelo.predict_proba(X_sleap_scaled)[:, 1]  # Probabilidad de que sea "Evento"

    # Agregar las predicciones y probabilidades al DataFrame
    df_sleap['Predicción'] = y_pred
    df_sleap['Confianza_Evento'] = y_prob

    # Comprobar si la columna 'Observation' existe en el DataFrame
    if 'Observation' in df_sleap.columns:
        print("\nEvaluación de Predicciones en CSV Anotado:")
        print(df_sleap[['Observation', 'Predicción', 'Confianza_Evento']].head(10))
    else:
        print("\nPredicciones realizadas en CSV no anotado (detección original de SLEAP):")
        print(df_sleap[['frame_idx', 'Predicción', 'Confianza_Evento']].head(10))

    # Visualización de la confianza a lo largo del tiempo
    plt.figure(figsize=(14, 6))
    colores = np.where(df_sleap['Predicción'] == 'Evento', 'red', 'blue')
    plt.scatter(df_sleap['frame_idx'], df_sleap['Confianza_Evento'], c=colores, label='Confianza de Evento', alpha=0.6)
    plt.axhline(y=0.5, color='r', linestyle='--', label='Umbral de Decisión')
    plt.xlabel('Tiempo (Frames)')
    plt.ylabel('Confianza')
    plt.title('Confianza del Modelo para Evento a lo largo del Tiempo')
    plt.legend()
    plt.show()

    # Retornar el DataFrame con predicciones
    return df_sleap

# Funciones auxiliares ya definidas anteriormente
def normalizar_coordenadas(df):
    width = 520  # Ancho del video en píxeles
    height = 245  # Alto del video en píxeles
    cols_x = [col for col in df.columns if '.x' in col]
    cols_y = [col for col in df.columns if '.y' in col]
    df[cols_x] = df[cols_x] / width
    df[cols_y] = df[cols_y] / height
    return df

def calcular_distancia(x1, y1, x2, y2):
    return np.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

def calcular_angulo(df):
    vector1 = np.column_stack((df['Cabeza.x'] - df['Lomo.x'], df['Cabeza.y'] - df['Lomo.y']))
    vector2 = np.column_stack((df['Base.x'] - df['Lomo.x'], df['Base.y'] - df['Lomo.y']))
    norm1 = np.linalg.norm(vector1, axis=1)
    norm2 = np.linalg.norm(vector2, axis=1)
    dot_product = np.einsum('ij,ij->i', vector1, vector2)
    valid_indices = (norm1 != 0) & (norm2 != 0)
    angles = np.full_like(norm1, fill_value=np.nan)  # Inicializar con NaN
    angles[valid_indices] = np.arccos(dot_product[valid_indices] / (norm1[valid_indices] * norm2[valid_indices]))
    return np.degrees(angles)

def calcular_angulo_hisopo(df):
    vector_cuerpo = np.column_stack((df['Base.x'] - df['Cabeza.x'], df['Base.y'] - df['Cabeza.y']))
    vector_hisopo = np.column_stack((270 - df['Lomo.x'], 50 - df['Lomo.y']))
    norm_cuerpo = np.linalg.norm(vector_cuerpo, axis=1)
    norm_hisopo = np.linalg.norm(vector_hisopo, axis=1)
    dot_product = np.einsum('ij,ij->i', vector_cuerpo, vector_hisopo)
    valid_indices = (norm_cuerpo != 0) & (norm_hisopo != 0)
    ang_hisopo = np.full_like(norm_cuerpo, fill_value=np.nan)  # Inicializar con NaN
    ang_hisopo[valid_indices] = np.arccos(dot_product[valid_indices] / (norm_cuerpo[valid_indices] * norm_hisopo[valid_indices]))
    return np.degrees(ang_hisopo)

def calcular_vel_acel(df, window_size=60):
    df['vel_nariz'] = calcular_distancia(
        df['Nariz.x'].shift(window_size), df['Nariz.y'].shift(window_size), df['Nariz.x'], df['Nariz.y']
    ) / window_size
    df['acel_nariz'] = df['vel_nariz'].diff(window_size) / window_size
    df['vel_ang_cuerpo'] = df['ang_cuerpo'].diff(window_size) / window_size
    df['vel_ang_hisopo'] = df['ang_hisopo'].diff(window_size) / window_size
    df[['vel_nariz', 'acel_nariz', 'vel_ang_cuerpo', 'vel_ang_hisopo']] = df[
        ['vel_nariz', 'acel_nariz', 'vel_ang_cuerpo', 'vel_ang_hisopo']
    ].fillna(0)
    return df

def extraer_caracteristicas(df, window_size=1):
    df['nariz_hisopo'] = calcular_distancia(df['Nariz.x'], df['Nariz.y'], 270, 50)
    df['cabeza_hisopo'] = calcular_distancia(df['Cabeza.x'], df['Cabeza.y'], 270, 50)
    df['long_cuerpo'] = calcular_distancia(df['Cabeza.x'], df['Cabeza.y'], df['Base.x'], df['Base.y'])
    df['ang_cuerpo'] = calcular_angulo(df)
    df['ang_hisopo'] = calcular_angulo_hisopo(df)
    df = calcular_vel_acel(df, window_size)
    df.dropna(inplace=True)
    return df

def filtrar_frames_habituacion(df, num_frames_habituacion=0):
    df_filtrado = df[df['frame_idx'] > num_frames_habituacion]
    return df_filtrado

# Ejemplo de uso de la función

rf_model_trained = 'modelo_RandomForest_A7-B4-C3.joblib'

sleap_csv_path_an = 'RawData/Rela_cropped_videos.002_C-3 crop.analysis.csv'

save_cluster_name = 'RawData/C-3 crop'

df_resultado = cargar_y_predecir_sleap(rf_model_trained, sleap_csv_path_an )

from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd

# Función para agrupar eventos cercanos usando DBSCAN
def agrupar_eventos(df, distancia_temporal=15, min_eventos=3):
    """
    Agrupa eventos cercanos temporalmente en clústeres usando DBSCAN.
    
    :param df: DataFrame con las predicciones y tiempos de los eventos.
    :param distancia_temporal: Número de frames para considerar eventos como cercanos.
    :param min_eventos: Número mínimo de eventos en un clúster para ser considerado como tal.
    :return: DataFrame con etiquetas de clúster asignadas.
    """
    # Filtrar solo los frames con predicción de "Evento"
    df_eventos = df[df['Predicción'] == 'Evento'].copy()

    # Revisar si hay eventos detectados
    if df_eventos.empty:
        print("No se encontraron eventos para agrupar.")
        df['Cluster'] = -1  # Sin clúster
        return df

    # Aplicar DBSCAN sobre los índices de los frames para detectar clústeres temporales
    X = df_eventos[['frame_idx']].values
    dbscan = DBSCAN(eps=distancia_temporal, min_samples=min_eventos, metric='euclidean')
    df_eventos['Cluster'] = dbscan.fit_predict(X)

    # Combinar la información de clústeres de eventos al DataFrame original
    df = df.merge(df_eventos[['frame_idx', 'Cluster']], on='frame_idx', how='left')
    df['Cluster'].fillna(-1, inplace=True)  # Etiqueta de -1 para frames sin clúster
    
    print("\nResumen de clústeres detectados:")
    print(df['Cluster'].value_counts())

    return df

# Aplicar la función para agrupar eventos
df_resultado_clustering = agrupar_eventos(df_resultado, distancia_temporal=30, min_eventos=3)

# Visualización de los resultados
plt.figure(figsize=(14, 6))

# Convertir los valores de los clústeres a números para que puedan ser interpretados por la colormap
# Asignar -1 a ruido y mapear los demás clústeres a un rango de colores
colores = df_resultado_clustering['Cluster'].astype(float)  # Convertir a float para manejar -1 correctamente

# Graficar los puntos con colores basados en los clústeres asignados
scatter = plt.scatter(
    df_resultado_clustering['frame_idx'], 
    df_resultado_clustering['Confianza_Evento'], 
    c=colores, cmap='inferno', alpha=0.5, s=15
)

# Línea horizontal de umbral
plt.axhline(y=0.5, color='r', linestyle='--', label='Umbral de Decisión')

# Etiquetas y leyenda
plt.xlabel('Tiempo (Frames)')
plt.ylabel('Confianza')
plt.title('Eventos y Clústeres Detectados')
plt.colorbar(scatter, label='Clúster ID')
plt.legend()
plt.savefig(f'{save_cluster_name}_cluster.png')
plt.show()


df_resultado_clustering.to_csv(f'{save_cluster_name}_cluster.csv', index=False)


````
